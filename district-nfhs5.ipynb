# National Family Health Survey, India - NFHS-5 

### Web Scraping  -  District level

```python
from bs4 import BeautifulSoup
import requests
import PyPDF2
import io
import re
import csv
```

```python
def get_urls(url):
    pdf_list = []
    
    # Get HTML file from the given url
    html_doc = requests.get(url)
    
    # check the status of getting HTML file from url
    if html_doc.status_code == 200:
        # Create a beautiful soup object for the HTML file to manipulate and extract data from it.
        soup = BeautifulSoup(html_doc.text, 'html.parser')
        
        dropdown = soup.find("select")
        dropdown_values = dropdown.find_all(target="new")

        for i in dropdown_values:    
            pdf_list.append('http://rchiips.org/nfhs/'+ i['value'])
    
    return pdf_list
```

```python
url = "http://rchiips.org/nfhs/districtfactsheet_NFHS-5.shtml"
state_urls = get_urls(url)
```

```python
type(state_urls)
```

```python
all_district_urls = []
for i in state_urls:   
    state_districts_list = get_urls(i)
    all_district_urls.append(state_districts_list)
```

```python
no_of_districts = 0
for i in all_district_urls:
    no_of_districts = no_of_districts + len(i)
    
    
no_of_districts
```
